```python
x = ratings_['userId'].value_counts() > 200
```

```python
y = x[x].index
```

```python
ratings = ratings_[ratings_['userId'].isin(y)]
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
rating_movies = ratings.merge(movies_, on="movieId")
```

```python
rating_movies.head()
```

```python
rating_movies.shape
```

```python
num_rating = rating_movies.groupby('title')['rating'].count().reset_index()
```

```python
num_rating.head()
```

```python
num_rating.rename(columns={'rating': 'num_of_rating'}, inplace=True)
```

```python
num_rating.head()
```

```python
final_rating = rating_movies.merge(num_rating, on='title')
```

```python
final_rating.head(2)
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['num_of_rating'] >= 50]
```

```python
final_rating.head()
```

```python
final_rating.sample(6)
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['userId', 'title'], inplace=True)
```

```python
final_rating.shape
```

```python
movies_pivot = final_rating.pivot_table(columns='userId', index='title', values='rating')
```

```python
movies_pivot
```

```python
movies_pivot.shape
```

```python
movies_pivot = movies_pivot.fillna(0)
```

```python
movies_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
movies_sparse = csr_matrix(movies_pivot)
```

```python
movies_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
# Creamos una instancia del modelo Nearest Neighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(movies_sparse)
```

```python
distance, suggestion = model.kneighbors(movies_pivot.iloc[28,:].values.reshape(1,-1), n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(movies_pivot.index[suggestion[i]])
```

```python
import pickle
pickle.dump(model, open('data_trained/model.pkl', 'wb'))
pickle.dump(movies_name, open('data_trained/movies_name.pkl', 'wb'))
pickle.dump(final_rating, open('data_trained/final_rating.pkl', 'wb'))
pickle.dump(movies_pivot, open('data_trained/movies_pivot.pkl', 'wb'))
```

```python
def recommend_movie(movie_name, movies_pivot, model):
    movie_id = np.where(movies_pivot.index == movie_name)[0][0]
    distance, suggestion = model.kneighbors(movies_pivot.iloc[movie_id, :].values.reshape(1, -1), n_neighbors=6)
    for distance_in_suggestion in range(len(suggestion)):
        movie = movies_pivot.index[suggestion[distance_in_suggestion]]
        for name in movie:
            print(name)
```

```python
movie_name = "Jumanji (1995)"
recommend_movie(movie_name, movies_pivot, model)
```
